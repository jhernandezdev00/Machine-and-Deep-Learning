##### PRACTICA No. 4. Redes Neuronales(SCIKIT-LEARN_MLPClassifier)
En esta práctica se implementará un modelo de clasificación una red neuronal con la técnica de propagación hacia atras (Por sus siglas en el idioma inglés NNBP). La red NNBP se implementará utilizando la libreria de código abierto SCIKIT-LEARN utilizando la clase MLPClassifier() 

Paso 1: Importar las librerías necesarias

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

Paso 2: Cargamos el dataset

```python
df_flor = pd.read_csv('../data/iris.csv', delimiter=";", decimal=".")
print(df_flor.head())
print(f"\nNumero de Muestras: {df_flor.shape[0]}\nNumero de Variables: {df_flor.shape[1]}")
```

PASO 3: Extraemos las variables predictoras.

```python
x = df_flor.iloc[:,:4]
print(x[0:3])
```

PASO 4: Ahora separamos la variable a predecir

```python
y = df_flor.iloc[:,4:]
print(y.head())
```

PASO 5: Separamos los datos con el 70% de los datos para entrenamiento y el 30% para testing

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y, train_size=0.7, random_state=0)
```

PASO 6: Mediante el constructor inicializa la instancia de red
Nota: Por defecto se usa 100 nodos en cada capa oculta, "solver" es el método de optimización

```python
MLP_Classifier = MLPClassifier(learning_rate_init=0.01, max_iter=500)
print(MLP_Classifier)
```

PASO 7: Entrenamos el modelo llamando al método fit

```python
MLP_Classifier.fit(xTrain,yTrain.iloc[:,0].values)
```

Paso 8: Imprimimos las predicciones en testing

```python
print("Las predicciones en Testing son: \n")

valores_Predict = MLP_Classifier.predict(xTest)

print(f"Valor Real \t => Valor Predicho ")
print("=================================")

for i in range(len(xTest)):
    print(f"{yTest.values[i]} \t => {valores_Predict[i]} ")
```

PASO 9: Calculamos los indices de calidad de la predicción

```python
def indice_general(MC, nombres = None):
    precision_global = np.sum(MC.diagonal())/ np.sum(MC)
    error_global = 1 - precision_global
    precision_categorica = pd.DataFrame(MC.diagonal()/np.sum(MC,axis=1)).T
    if nombres != None:
        precision_categorica.columns = nombres
    return {"Matriz de Confusion:": MC,
            "Precisión Global": precision_global,
            "Error Global": error_global,
            "Precisión por Categoria": precision_categorica}
```

```python
MC = confusion_matrix(yTest, valores_Predict)
indices = indice_general(MC, list(np.unique(y)))

for k in indices:
    print("\n%s: \n%s"%(k,str(indices[k])))
```
