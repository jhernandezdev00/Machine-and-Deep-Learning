```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

PASO 1: Importamos los datos del archivo con los datos de la flores, indicandole que tome como cabezera la primera linea que separe cada dato donde encuentre un punto y coma y que solo tome las primeras 100 columnas, debido a que solo 
trabajaremos con los dos primeros tipos de flores

```python
df_iris = pd.read_csv('../Data/iris.csv',header=0,sep=';',nrows=100)
print(df_iris)
```

PASO 2: Separamos los datos que necesitamos en arrays independientes para manipularlos

```python
longSepalo = np.array(df_iris['s.largo'])
longPetalo = np.array(df_iris['p.largo'])
especie = np.array(df_iris['tipo'])
```

```python
print("Longitud del sépalo:", longSepalo[:5])
print("Longitud del pétalo:", longPetalo[:5])
print("Especie:", especie[:5])
```

PASO 3: Separaremos nuestros datos para utilizar una parte para pruebas y la otra para el entrenamiento 


```python
longSepalo_train, longSepalo_test, longPetalo_train, longPetalo_test, especie_train, especie_test= train_test_split(longSepalo,longPetalo,especie,test_size=0.2)

print("Longitud del sépalo:", longSepalo_train[:5])
print("Longitud del pétalo:", longPetalo_train[:5])
print("Especie:", especie_train[:5])
```

PASO 4: Modelamos nuestra red

```python
# Funcion de activacion de la Neurona
def activacion(peso1,peso2,petaloX,sepaloY,bias):
    z = (peso1*petaloX)+(peso2*sepaloY)
    if z.sum()+bias > 0:
        return 1
    else:
        return 0

print("\n ====== Inicio Del Entrenamiento ====== \n")

# Funcion de Entrenamiento
peso1 = np.random.uniform(-1,1,size=2)
peso2 = np.random.uniform(-1,1,size=2)
bias = np.random.uniform(-1,1)
tasa_de_aprendizaje = 0.01
epocas = 100

for epocas in range(epocas):
    error_total=0
    for i in range(len(longSepalo_train)):
        resultado = activacion(peso1, peso2,longPetalo_train[i],longSepalo_train[i],bias)
        if especie_train[i] == 'setosa' :
            estatus_especie = 0
        else:
            estatus_especie = 1
        
        error = estatus_especie - resultado
        error_total += error**2
        peso1 += tasa_de_aprendizaje*longPetalo_train[i]*error
        peso2 += tasa_de_aprendizaje*longSepalo_train[i]*error
        bias += tasa_de_aprendizaje*error

        print(error_total, end=" ")
    #print("==============================================",end="/n")

print("\n ====== Fin Del Entrenamiento ====== \n")
```

```python
predicciones = []

# Funcion de Pruebas
print("\n ====== Inicio De las Pruebas ====== \n")
for i in range(len(longSepalo_test)):
     result1 = activacion(peso1, peso2,longPetalo_test[i],longSepalo_test[i],bias)
     if result1 == 0 :
             predicciones.append(0)
             print("[",longPetalo_test[i],longSepalo_test[i],"] --> Setosa")
     else:
            predicciones.append(1)
            print("[",longPetalo_test[i],longSepalo_test[i],"] --> Versicolor")

print("\n ====== Finalizacion De las Pruebas ====== \n")    
        
print("====== Comparacion de Resultados Esperados ====== \n")
print(especie_test) 
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
```

```python
etiquetas_reales = np.array([0 if especie == "setosa" else 1 for especie in especie_test])
etiquetas_predichas = predicciones

matriz_confusion = confusion_matrix(etiquetas_reales, etiquetas_predichas)

print("\n🔹 Matriz de Confusión:")
print(matriz_confusion)

# Reporte de clasificación
print("\n🔹 Reporte de Clasificación:")
print(classification_report(etiquetas_reales, etiquetas_predichas, target_names=["Setosa", "Versicolor"]))
```
