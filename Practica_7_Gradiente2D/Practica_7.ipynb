PRACTICA No. 5. Descenso del Gradiente 2D


La idea principal es representar el error como una función continua de los pesos e intentar ilevar la red hacia una configuración de valores de activación que nos permita encontrar el mínimo de esta
función.

Para determinar la magnitud y dirección del cambio que debemos introducir en el vector de pesos para reducir el error que está cometiendo, utilizaremos el concepto de «pendiente», que indica que la pendiente de un punto dado es el gradiente de la tangente a la curva de la función en el punto dado.

El gradiente es la derivada de una operación tensorial. Es la generalización del concepto de derivadas a funciones que toman fensores como entradas

Paso 1: Importar las librerías necesarias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Paso 2: Definimos la funcion $$f(x) = 2x^2 * cos(x) -5x$$

```python
def funcion(x):
    return 2*(pow(x,2)) * np.cos(x) - 5 * x
```

Paso 3: Graficamos la funcion $$f(x) = 2x^2 * cos(x) -5x$$

```python
t1 = np.arange(-5.0,5.0,0.1)

plt.plot(t1, funcion(t1))
plt.show()
```

PASO 4: IMPLEMENTAMOS EL ALGORITMO DEL DESCENSO DEL GRADIENTE

PASO 4_1: Tomar un punto al azar de X0

```python
t1 = np.arange(-5.0,5.0,0.1)
plt.plot(t1, funcion(t1))

point_x0 = [np.random.randint(-5,5)]
point_y0 = [funcion(point_x0[0])]

print(f"POS X: {point_x0[0]}")
print(f"POS Y: {point_y0[0]}")


plt.plot(point_x0,point_y0,'ro')

plt.show()
```

PASO 4_2: Calcular el valor de la pendiente (Primera Derivada) de f(x) $$-2x^2sin(x)+4xcos(x) -5 $$

```python
def funcion_primera_derivada(x):
    return (-2*pow(x,2)*np.sin(x) + 4*x*np.cos(x) - 5)
```

```python
pendiente = funcion_primera_derivada(point_x0[0])
print(f"Pendiente: {pendiente}")
```

PASO 4_3: Camine en direccion opuesta a la pendiente 
$$x_1 = x_0 - eta * f'(x_0)$$

NOTA: ETA representa la Taza de Aprendizaje

```python
eta = 0.05

point_x0.append(point_x0[0] - eta * pendiente)
point_y0.append(funcion(point_x0[0] - eta * pendiente))

print(f"POS X: {point_x0}")
print(f"POS Y: {point_y0}")
```

PASO 4_4: Graficamos el punto 2

```python
t1 = np.arange(-5.0,5.0,0.1)
plt.plot(t1, funcion(t1))

plt.plot(point_x0,point_y0,'ro')

plt.show()
```

PASO 4_5: Graficamos el punto 3

```python
point_x0.append(point_x0[1] - eta * pendiente)
point_y0.append(funcion(point_x0[1] - eta * pendiente))

print(f"POS X: {point_x0}")
print(f"POS Y: {point_y0}")

print(f"Pendiente 0: {funcion_primera_derivada(point_x0[0])}")
print(f"Pendiente 1: {funcion_primera_derivada(point_x0[1])}")
print(f"Pendiente 2: {funcion_primera_derivada(point_x0[2])}")

t1 = np.arange(-5.0,5.0,0.1)
plt.plot(t1, funcion(t1))

plt.plot(point_x0,point_y0,'ro')

plt.show()
```

#### PROCESO COMPLETO

```python
point_x0 = [np.random.randint(-5,5)]
point_y0 = [funcion(point_x0[0])]

print(f"POS X: {point_x0[0]}")
print(f"POS Y: {point_y0[0]}")
```

```python
epsilon = 1e-6
eta = 0.05
max_iter = 200

point_x0 = [0]
point_y0 = [funcion(point_x0[0])]

for i in range(max_iter):
    pendiente = funcion_primera_derivada(point_x0[i])  # Cálculo del gradiente
    nuevo_x = point_x0[i] - eta * pendiente
    nuevo_y = funcion(nuevo_x)

    if abs(nuevo_x - point_x0[i]) < epsilon:
        print(f"Se detiene en iteración {i}, cambio menor a {epsilon}")
        break

    point_x0.append(nuevo_x)
    point_y0.append(nuevo_y)
    
    print(f"Iteración {i}: x = {nuevo_x}, y = {nuevo_y}, pendiente = {pendiente}")
```

```python
t1 = np.arange(-5.0,5.0,0.1)

fig, ax = plt.subplots()
ax.plot(t1, funcion(t1))

plt.plot(point_x0,point_y0,'ro')
plt.plot(point_x0[-1], point_y0[-1], 'bo', markersize=8, label="Último punto")

plt.legend()
plt.show()
```
